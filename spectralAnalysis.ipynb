### Library Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading File

```python
astro = pd.read_fwf('ASTRO_0001_00001.sed',skiprows = 26, sep = '\t', header = None)
```

### Extracting Data

```python
column0 = astro[0].str.split('\t')[1:] 
column1 = astro[1].str.split('\t')[1:]
```

```python
wave_lngth = []
reflectance = []
for i in range(len(column0.to_numpy())):
    lngth = column0.to_numpy()[i][0]
    rfl = column1.to_numpy()[i][1] 
    wave_lngth.append(lngth) 
    reflectance.append(rfl)
```

### Plotting Data

```python
wave_lngth = np.array(wave_lngth).astype('float') 
reflectance = np.array(reflectance).astype('float')
```

```python
plt.plot(wave_lngth, reflectance)
plt.title('Spectral Reflectance Curve')
plt.xlabel('Wave Length')
plt.ylabel('Reflectance')
```
