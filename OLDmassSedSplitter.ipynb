```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
inputDir = "Z:\\vulcano\\Vulcano\\16062019\\astrobay\\small_samples\\Astro_Sample1\\"
extType = ".sed"
sourceName = "ASTRO_0001_00001"
output_dir = "Z:\\vulcano\\output\\" + sourceName + "\\"
sourceTarget = inputDir + sourceName + extType
Output_UV = sourceName + "_UV" + extType
Output_VIS = sourceName + "_VIS" + extType
Output_IR = sourceName + "_IR" + extType
outputTarget_UV = output_dir + Output_UV
outputTarget_VIS = output_dir + Output_VIS
outputTarget_IR = output_dir + Output_IR

##wavelengths in nm
WL_MIN = 350.0
WL_UV_Cutoff = 400.0
WL_IR_Cutoff = 700.0
WL_MAX = 2500.0
```

```python
##open source, create target files
st = open(sourceTarget, "r")

#t1.write(header)

file = st.read()

wlr_index = file.find("Wavelength Range:") + len("Wavelength Range:")
wlr_index_end = file.find("Latitude:")

Channels_index =  file.find("Channels: ") + len("Channels: ")
Channels_index_end = file.find("Columns ")

header_end_index = file.find("Reflect. [1.0]\n") + len("Reflect. [1.0]\n")
data = file[header_end_index:len(file)]


header_UV = file[0:wlr_index] + str(WL_MIN) + "," + str(WL_UV_Cutoff) + "\n"
header_VIS = file[0:wlr_index] + str(WL_UV_Cutoff) + "," + str(WL_IR_Cutoff) + "\n"
header_IR = file[0:wlr_index] + str(WL_IR_Cutoff) + "," + str(WL_MAX) + "\n"

header_UV += file[wlr_index_end:Channels_index] + str(int(WL_UV_Cutoff) - int(WL_MIN)) + "\n"
header_VIS += file[wlr_index_end:Channels_index] + str(int(WL_IR_Cutoff) - int(WL_UV_Cutoff)) + "\n"
header_IR += file[wlr_index_end:Channels_index] + str(int(WL_MAX) - int(WL_IR_Cutoff)) + "\n"

header_UV += file[Channels_index_end:header_end_index]
header_VIS += file[Channels_index_end:header_end_index]
header_IR += file[Channels_index_end:header_end_index]

header_UV = header_UV.replace(extType, "_UV" + extType)
header_VIS = header_VIS.replace(extType, "_VIS" + extType)
header_IR = header_IR.replace(extType, "_IR" + extType)

data_UV = header_UV
data_VIS = header_VIS
data_IR = header_IR

while len(data) > 4:
    number_start = 1-data[0:1].isalnum()
    number_end = data.find("\t")
    line_end = data.find("\n")+1
    number = float(data[number_start:number_end])
    dataLine = data[0:line_end]

    if (number >= WL_UV_Cutoff) and (number <= WL_IR_Cutoff):
        print(str(number) + "nm is in the visible Spectrum!")
        #t_VIS.write(dataLine)
        data_VIS += dataLine
    
    elif number <= WL_UV_Cutoff:
        print(str(number) + "nm is in the UV Spectrum!")
        #t_UV.write(dataLine)    
        data_UV += dataLine
        
    elif number >= WL_IR_Cutoff:
        print(str(number) + "nm is in the IR Spectrum!")
        #t_IR.write(dataLine)    
        data_IR += dataLine
        
    data = data.replace(dataLine, "")
    
print("Data partitioning complete!")


# to write directory. commented out since it only needs to be written once
##os.mkdir(output_dir)


# Writing files!
t_UV = open(outputTarget_UV, "w")
t_UV.write(data_UV)

t_VIS = open(outputTarget_VIS, "w")
t_VIS.write(data_VIS)

t_IR = open(outputTarget_IR, "w")
t_IR.write(data_IR)




print("Files created!")
```
